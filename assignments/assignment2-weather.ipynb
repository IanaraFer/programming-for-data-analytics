```python
url = 'https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv'
data = pd.read_csv(url)

print(data.head())
```

#### I was confused about what time to use. Them I had decided to use all but in different plot to see if have any diffenrence.

```python
# Convert 'Date' column to datetime format using "createddatatime"
data['createdDateTime'] = pd.to_datetime(data['createdDateTime'])

# Plotting
plt.figure(figsize=(10,5))
plt.plot(data['createdDateTime'], data['dryBulbTemperature_Celsius'], marker='o')
plt.title('TdryBulbTemperature_Celsius Over Time')
plt.xlabel('Time')
plt.ylabel('dryBulbTemperature_Celsius')
plt.grid(True)
plt.show()
```

```python
# Convert 'Date' column to datetime format using "localReportEndDateTime"
data['localReportEndDateTime'] = pd.to_datetime(data['localReportEndDateTime'])

# Plotting
plt.figure(figsize=(10,5))
plt.plot(data['localReportEndDateTime'], data['dryBulbTemperature_Celsius'], marker='o')
plt.title('TdryBulbTemperature_Celsius Over Time')
plt.xlabel('Time')
plt.ylabel('dryBulbTemperature_Celsius')
plt.grid(True)
plt.show()
```
